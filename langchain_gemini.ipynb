```python
from langchain.document_loaders.csv_loader import CSVLoader
loader=CSVLoader(file_path='/content/drive/MyDrive/Generative AI/codebasics_faqs.csv',source_column='prompt', encoding='latin-1')
data=loader.load()
data
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embeddings=GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-exp-03-07")
text_id=embeddings.embed_query("i try to learn langchain")
text_id
```

```python
len(text_id),type(text_id)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
embeddings=HuggingFaceEmbeddings()
vectorDB=FAISS.from_documents(documents=data,embedding=embeddings)
```

```python
retriever=vectorDB.as_retriever()
reT=retriever.get_relevant_documents("for how long is this course vaild")
reT
```

```python
from langchain.chains import RetrievalQA
chain=RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriever,return_source_documents=True,input_key="query")
```

```python
res=chain('It looks like the fact_sales_monthly table is missing. Can you share why it\'s currently unavailable?')
```

```python
from langchain.prompts import PromptTemplate

prompt_template = """Given the following context and a question, generate an answer based on this context only.
In the answer try to provide as much text as possible from "response" section in the source document context without making much changes.
If the answer is not found in the context, kindly state "I don't know." Don't try to make up an answer.

CONTEXT: {context}

QUESTION: {question}"""


PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
chain_type_kwargs = {"prompt": PROMPT}


from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(llm=llm,
                            chain_type="stuff",
                            retriever=retriever,
                            input_key="query",
                            return_source_documents=True,
                            chain_type_kwargs=chain_type_kwargs)
```

```python
chain('Do you provide job assistance and also do you provide job gurantee?')
```
